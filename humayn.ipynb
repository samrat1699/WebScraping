```python
final = pd.DataFrame()


for page in range(1,8):
    webpage=requests.get('https://www.rokomari.com/book/author/1/humayun-ahmed?ref=mm_p0&page={}'.format(page)).text
    soup = BeautifulSoup(webpage,'html.parser')
   
    company=soup.find_all('div',class_='book-list-wrapper')
    books=[]
    booksA=[]
    Price=[]
    disPrice=[]
    
    
    for i in company:
        try:
            books.append(i.find('p', class_='book-title').text.strip())
        except:
            books.append(np.nan)
            
        try:
            booksA.append(i.find('p', class_='book-author').text.strip())
        except:
            booksA.append(np.nan)
            
        try:
            Price.append(i.find('strike', class_='original-price pl-2').text.strip()) 
        except:
            Price.append(np.nan)
            
        try:
            disPrice.append(i.find('span', class_="").text.strip()) 
        except:
            disPrice.append(np.nan)     
    
    
        
df=pd.DataFrame({'books':books,
        'booksA':booksA,
        'Price':Price,
        'disPrice':disPrice,
                    })

final=final.append(df,ignore_index=True)
       
```

```python
final.to_csv("Humayon9.csv")
```
